```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.animation as animation
from IPython.core.display import HTML
from tkinter import *
import tkinter as tk
import os
import imageio
from PIL import Image, ImageTk
plt.rcParams.update({'figure.max_open_warning': 0})

global size1
global size2
global inf
global isol1
global isol2
global ech
global vac1
global vac2
global douane

root = Tk()
def fin():
    temp = Toplevel(root, width = 2000)
    confirmation = Label(temp, text = "Confirmez vous ces paramètres ?")
    oui = Button(temp, text = "Valider", command = temp.destroy and root.destroy)
    non = Button(temp, text = "Retour", command = temp.destroy)
    confirmation.grid(row = 0, column = 0, columnspan = 2) 
    oui.grid(row = 1, column = 0)
    non.grid(row = 1, column = 1)

b = Button(root, text="Valider", command=fin, justify = RIGHT) 
b.grid(row = 10, column = 4)


valueTQ1 = IntVar(root)
valueTQ2 = IntVar(root)
scaleTQ1 = Scale(root, from_=0, to=50,showvalue=True, label='Sevérité de la quarantaine zone 1',
variable=valueTQ1, tickinterval=20, orient='h',length = 200)
scaleTQ2 = Scale(root, from_=0, to=50,showvalue=True, label='Sevérité de la quarantaine zone 2',
variable=valueTQ2, tickinterval=20, orient='h',length = 200)
entryTQ1 = Entry(root, textvariable=valueTQ1)
entryTQ2 = Entry(root, textvariable=valueTQ2)
scaleTQ1.grid(row= 3, column = 0)
scaleTQ2.grid(row= 3, column = 1)
entryTQ1.grid(row= 4, column = 0)
entryTQ2.grid(row= 4, column = 1)

valueD = IntVar(root)
entryD = Entry(root, textvariable=valueD)
scaleD = Scale(root, from_=0, to=100,showvalue=True, label='Sevérité de la douane',
variable=valueD, tickinterval=20, orient='h',length = 200)
scaleD.grid(row = 3, column = 2)
entryD.grid(row= 4,column = 2)

valueV1 = IntVar(root)
entryV1 = Entry(root, textvariable=valueV1)
scaleV1 = Scale(root, from_=0, to=100,showvalue=True, label='Pourcentage de personnes vaccinées zone 1',
variable=valueV1, tickinterval=20, orient='h',length = 200)
scaleV1.grid(row= 6, column = 0)
entryV1.grid(row= 7, column = 0)
valueV2 = IntVar(root)
entryV2 = Entry(root, textvariable=valueV2)
scaleV2 = Scale(root, from_=0, to=100,showvalue=True, label='Pourcentage de personnes vaccinées zone 2',
variable=valueV2, tickinterval=20, orient='h',length = 200)
scaleV2.grid(row= 6, column = 1)
entryV2.grid(row= 7, column = 1)

valueT1 = IntVar(root)
entryT1 = Entry(root, textvariable=valueT1)
scaleT1 = Scale(root, from_=5, to=50,showvalue=True, label='Taille zone 1',
variable=valueT1, tickinterval=10, orient='h',length = 200,)
scaleT1.grid(row= 8, column = 0)
entryT1.grid(row= 9, column = 0)
valueT2 = IntVar(root)
entryT2 = Entry(root, textvariable=valueT2)
scaleT2 = Scale(root, from_=5, to=50,showvalue=True, label='Taille zone 2',
variable=valueT2, tickinterval=10, orient='h',length = 200)
scaleT2.grid(row= 8, column = 1)
entryT2.grid(row= 9, column = 1)

valueE = IntVar(root)
entryE = Entry(root, textvariable=valueE)
scaleE = Scale(root, from_=0, to=100,showvalue=True, label="Taux d'échange",
variable=valueE, tickinterval=20, orient='h',length = 200)
scaleE.grid(row = 6, column = 2)
entryE.grid(row= 7,column = 2)

valueI = IntVar(root)
entryI = Entry(root, textvariable=valueI)
scaleI = Scale(root, from_=10, to=100,showvalue=True, label="Taux d'infection",
variable=valueI, tickinterval=20, orient='h',length = 200)
scaleI.grid(row = 8, column = 2)
entryI.grid(row= 9,column = 2)

valueG = IntVar(root)
entryG = Entry(root, textvariable=valueG)
scaleG = Scale(root, from_=4, to=10,showvalue=True, label="Durée infection(jours)",
variable=valueG, tickinterval=20, orient='h',length = 200)
scaleG.grid(row = 10, column = 2)
entryG.grid(row= 11,column = 2)


root.mainloop()



size1 = valueT1.get()
size2 = valueT2.get()
inf = valueI.get()/100
isol1 = valueTQ1.get()/100
isol2 = valueTQ2.get()/100
ech = valueE.get()/100
vac1 = valueV1.get()/100
vac2 = valueV2.get()/100
douane = valueD.get()/100
t_gue = valueG.get()


def init_map_s(taille):#Creation map sans infectés

    map = array([np.random.uniform(low = 0.0, high = 0.0, size = taille) for i in range(taille)])
    return map

def init_map(taille): #Creation map avec un infecté
    map = array([np.random.uniform(low = 0.0, high = 0.0, size = taille) for i in range(taille)])
    map[np.random.choice(taille - 1)][np.random.choice(taille - 1)] = 1
    return map

def vaccination(map, size, vac):
    i = 0
    j = 0
    while i < size:
        while j < size:
            if np.random.binomial(1, vac, 1) == 1 and map[i][j] != 1:
                map[i][j] = -50
            j = j + 1
        j = 0
        i = i + 1
    return map

def evolution(zone,taille):
    for i in range(taille):
        for j in range(taille):
            if zone[i][j] > 0:
                zone[i][j] = zone[i][j] + 1
                if zone[i][j] == t_gue:
                    zone[i][j] = -1
            elif zone[i][j] < -1 and zone[i][j] != -50:
                zone[i][j] = zone[i][j] - 1
                if zone[i][j] == t_gue * -1:
                    zone[i][j] = -1
    return zone

def infection(map,size,t_inf):
    for i in range(size):
        for j in range(size):
            if map[i][j] > 0:
                try:
                    if map[i + 1][j] == 0 and i != size - 1:
                        map[i + 1][j] = np.random.binomial(1, t_inf, 1)
                except:
                    pass
                try:
                    if map[i - 1][j] == 0 and i != 0:
                        map[i - 1][j] = np.random.binomial(1, t_inf, 1)
                except:
                    pass
                try:
                    if map[i][j + 1] == 0 and j != size - 1:
                         map[i][j + 1] = np.random.binomial(1, t_inf, 1)
                except:
                    pass
                try:
                    if map[i][j - 1] == 0 and j != 0:
                        map[i][j - 1] = np.random.binomial(1, t_inf, 1)
                except:
                    pass
    return map

def isolate(map,size,t_isol):
    for i in range(size):
        for j in range(size):
            if map[i][j] > 0:
                if np.random.binomial(1, t_isol, 1) == 1:
                    map[i][j] = map[i][j] * -1
    return map

def check(map,size):
    c = 1
    for i in range(size):
        for j in range(size):
            if map[i][j] != 0 and map[i][j] != -1 and map[i][j] != -50:
                c = 0
    return c

def echange(map,map2,size1,size2):
    if np.random.binomial(1, ech, 1) == 1:
        x1 = np.random.choice(size1 - 1)
        y1 = np.random.choice(size1 - 1)
        x2 = np.random.choice(size2 - 1)
        y2 = np.random.choice(size2 - 1)
        if ((douane > 0 and ((map[x1][y1] == 0 or map[x1][y1] == -1) and (map2[x2][y2] == 0 or map2[x2][y2] == -1))) or douane == 0):
            if douane == 0 or np.random.binomial(1,douane,1) != 1:
                temp = map[x1][y1]
                map[x1][y1] = map2[x2][y2]
                map2[x2][y2] = temp
    return map,map2

def bilan(map,map2,size,size2):
    nb_inf = 0
    for i in range(size):
        for j in range(size):
            if map[i][j] == -1:
                nb_inf = nb_inf + 1
    for i in range(size2):
        for j in range(size2):
            if map2[i][j] == -1:
                nb_inf = nb_inf + 1
    return nb_inf

def main():
    tab = []
    tab2 = []
    i = 0
    steps = 0
    m = init_map(size1)
    m2 = init_map_s(size2)
    m = vaccination(m,size1,vac1)
    m2 = vaccination(m2,size2,vac2)
    tab.append(m)
    tab2.append(m2)
    while(check(m,size1) == 0 or check(m2,size2) == 0):
        m = evolution(m,size1)
        m = infection(m,size1,inf)
        m = isolate(m,size1,isol1)
        m2 = evolution(m2,size2)
        m2 = infection(m2,size2,inf)
        m2 = isolate(m2,size2,isol2)
        m,m2 = echange(m,m2,size1,size2)
        tab.append(m.copy())
        tab2.append(m2.copy())
        steps = steps + 1
    fig, ax = plt.subplots()

    cmap = mpl.colors.ListedColormap(['cyan', 'yellow', 'pink', 'green','red'])
    bounds = [-100,-50,-7,-1.1,-0.1,0.1,100]
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
    im = plt.imshow(tab[0], animated=True,interpolation = 'nearest',cmap = cmap,norm=norm)

    def updatefig(i):
        im.set_array(tab[i+1])
        return im

    ani = animation.FuncAnimation(fig, updatefig, frames=steps, interval=200)
    ani.save('test.gif', fps=3, writer="imagemagick", codec="libx264")


    fig2, ax2 = plt.subplots()
    cmap = mpl.colors.ListedColormap(['cyan', 'yellow', 'pink', 'green','red'])
    bounds = [-100,-50,-7,-1.1,-0.1,0.1,100]
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
    im2 = plt.imshow(tab2[0], animated=True,interpolation = 'nearest',cmap = cmap,norm=norm)
    def updatefig2(j):
            im2.set_array(tab2[j+1])
            return im2
    ani2 = animation.FuncAnimation(fig2, updatefig2, frames=steps, interval=200)
    ani2.save('test2.gif', fps=3, writer="imagemagick", codec="libx264")


    class GifAnimatedLabel(Label): #Classe qui n'est pas de nous https://openclassrooms.com/forum/sujet/gif-anime-dans-tkinter
        def __init__(self, master, filename, speed, *args, **kwargs):
            self.speed = speed
            self.frames = []
            i = 0
            while True:
                try:
                    p = tk.PhotoImage(file=filename, format ="gif - {}".format(i))
                except tk.TclError:
                    break
                self.frames.append(p)
                i += 1
            super().__init__(master, image = self.frames[0], *args, **kwargs)
            self.frame_idx = 0
            self.num_frames = i
            self.after(self.speed, self._animate)

        def _animate(self):
            self.frame_idx = (self.frame_idx + 1) % self.num_frames
            self['image'] = self.frames[self.frame_idx]
            self.after(self.speed, self._animate)

        
    app = tk.Tk()
    zone1_text = Label(app, text="Zone 1")
    zone1_text.grid(row = 0, column = 0)
    zone2_text = Label(app, text="Zone 2")
    zone2_text.grid(row = 0, column = 1)
    zone1 = GifAnimatedLabel(app, filename="test.gif", speed=500)
    zone1.grid(row = 1, column = 0)
    zone2 = GifAnimatedLabel(app, filename="test2.gif", speed=500)
    zone2.grid(row = 1, column = 1)
    canvas = Canvas(app, width=75, height=60, background='yellow')
    txt = canvas.create_text(37.5, 30, text="Individu isolé", font="Arial 9 italic", fill="black")
    canvas2 = Canvas(app, width=75, height=60, background='pink')
    txt2 = canvas2.create_text(37.5, 30, text="Individu \nimmunisé", font="Arial 9 italic", fill="black")
    canvas.grid(row = 2 , column = 0)
    canvas2.grid(row = 2 , column = 1)
    canvas3 = Canvas(app, width=75, height=60, background='green')
    txt3 = canvas3.create_text(37.5, 30, text="Individu sain", font="Arial 9 italic", fill="black")
    canvas4 = Canvas(app, width=75, height=60, background='cyan')
    txt4 = canvas4.create_text(37.5, 30, text="Individu \nvacciné", font="Arial 9 italic", fill="black")
    canvas3.grid(row = 3 , column = 0)
    canvas4.grid(row = 3 , column = 1)
    canvas5 = Canvas(app, width=75, height=60, background='red')
    txt5 = canvas5.create_text(37.5, 30, text="Individu \nmalade", font="Arial 9 italic", fill="black")
    canvas5.grid(row = 4 , column = 0)
    bil = Label(app, text="L'épidemie a duré "+ str(steps) + " jours et il y a eu " + str(bilan(m,m2,size1,size2)) + " infectés")
    bil.grid(row = 4, column = 1) 
    app.mainloop()

main()
y = input("Voulez vous recommencer avec les mêmes parametres ? O pour oui ")
while (y == 'O'):
    main()
    y = input("Voulez vous recommencer avec les mêmes parametres ? O pour oui ")
```
