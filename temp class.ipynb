```python
import csv
from pylab import *
import random
import math

def infection(pays):
    if pays.pops >= 0 or pays.popinf == 0:
        nb_inf = sum(np.random.binomial(pays.popinf, 0.3,1000))/1000
        nb_inf = int(floor(nb_inf))
        pays.popinf = pays.popinf + nb_inf
        pays.pops = pays.pops - nb_inf

class Pays:
    def __init__(self, name, population):
        self.name = name
        self.popt = population
        self.popinf = 100
        self.pops = self.popt - self.popinf
        self.popm = 0
        self.popimu = 0
    
    def __str__(self):
        return "Pays {}, population {}".format(self.name, self.popt)
    
    def __repr__(self):
        return "<{}>".format(str(self))

tab_pays = {}
with open('population_par_pays.csv','r') as userFile:
    userFileReader = csv.reader(userFile)
    next(userFileReader) # Saute la première ligne

    for row in userFileReader:
        try:
            tab_pays[row[0]] = Pays(row[0], int(row[-1]))
        except:
            pass
print(tab_pays["Aruba"].popinf)
infection(tab_pays["Aruba"])
print(tab_pays["Aruba"].popinf)
print(tab_pays["Aruba"].pops)
print(tab_pays["Aruba"].popt)
```
